```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Carrega uma imagem colorida
img = cv2.imread('cat.jpg')

# Obtm os canais R, G e B da imagem
B, G, R = cv2.split(img)

# Aplica a frmula para converter em tons de cinza
gray = (0.3 * R + 0.59 * G + 0.11 * B).astype(np.uint8)
# gray = ((0.3 * R +255)+ (0.59 * G + 255)+ (0.11 * B + 255)).astype(np.uint8)

plt.figure(figsize=(12, 8))
plt.imshow(gray)
plt.title('Imagens S')
plt.axis('off')  # No mostrar os eixos
plt.show()
```
